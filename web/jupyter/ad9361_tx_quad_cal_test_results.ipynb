```python
import telemetry
import matplotlib.pyplot as plt
```

```python
tel = telemetry.searches(server="alpine")
plt.clf()

for chan in range(2):
    x, y, t = tel.ad9361_tx_quad_cal_test(channel=chan)
    y = [y[k] / t[k] for k in range(len(y))]
    plt.plot(x, y)
    plt.ylabel('Quad Cal Failure %')
    plt.xlabel('Test Date')
```

```python
import plotly.graph_objects as go
import numpy as np
fig = go.Figure()
for chan in range(2):
    x, y, t = tel.ad9361_tx_quad_cal_test(channel=chan)
    y = [y[k] / t[k] for k in range(len(y))]
    fig.add_trace(go.Scatter(x=x, y=y, name="channel "+str(chan)))
fig.update_layout(xaxis_title = 'Test Date', yaxis_title='Quad Cal Failure %')
fig.show()
```
