```python
import telemetry
import matplotlib.pyplot as plt
```

```python
tel = telemetry.searches(server="alpine")
plt.clf()

for chan in range(2):
    x, y, t = tel.ad9361_tx_quad_cal_test(channel=chan)
    y = [y[k] / t[k] for k in range(len(y))]
    plt.plot(x, y)
    plt.ylabel('Quad Cal Failure %')
    plt.xlabel('Test Date')
```
